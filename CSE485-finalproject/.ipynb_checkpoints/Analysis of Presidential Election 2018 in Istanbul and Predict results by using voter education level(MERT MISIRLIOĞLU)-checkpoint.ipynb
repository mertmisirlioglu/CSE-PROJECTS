## Machine Learning 

I used three different algoritm for regression. First one is Linear Regression , second one is KNN regression , and last one is Neural Network Regressor. My goal is getting mean squared error less than 0.25

```python
X = secogr[['okuma_yazma_bilmeyen_sayisi', 'okuma_yazma_bilip_okul_bitirmeyen', 'ilkokul', 'ilköğretim','ortaokul','lise','yüksekokul_veya_fakülte','yüksek_lisans','doktora','bilinmeyen']] 
y = secogr[['Recep_Tayip_Erdoğan_oy_yüzdesi','Muharrem_İnce_oy_yüzdesi','Meral_Akşener_oy_yüzdesi']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
import pickle
loaded_linear = pickle.load(open('data\linearsec.sav', 'rb'))
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
y_pred = loaded_linear.predict(X_test)
test_error_linear = mean_squared_error(y_test,y_pred)
print(test_error_linear)
```

Linear Regression mean squared error is less than 0.25 , this is verify my hypothesis. (Proof 1)

```python
print("Predicted values : \n",y_pred)
print("Actual values : \n", np.array(y_test))

countys = ['Başakşehir', 'Esenyurt', 'Beyoğlu', 'Şile', 'Esenler', 'Güngören','Eyüpsultan', 'Maltepe', 'Bağcılar', 'Üsküdar']
plt.figure(num=None, figsize=(30, 10), dpi=80, facecolor='w', edgecolor='k')


y_pos = np.arange(len(countys))
truelabels = np.array(y_test)
plt.xticks(y_pos, countys)

line1 = plt.bar(y_pos-0.47 , truelabels[:,0], color = 'black', width = 0.15 , label='label1')
line2 =plt.bar(y_pos-0.32 , y_pred[:,0], color = 'orange', width = 0.15 , label='label2')
plt.bar(y_pos-0.15, truelabels[:,1], color = 'black', width = 0.15 )
line3 = plt.bar(y_pos, y_pred[:,1], color = 'b', width = 0.15, label='label3')
plt.bar(y_pos+0.17 , truelabels[:,2], color = 'black', width = 0.15)
line4 = plt.bar(y_pos+0.32 , y_pred[:,2], color = 'g', width = 0.15 , label='label4')
plt.title('Linear Regression predicts table')
plt.ylabel('Vote percentage(%)')
plt.xlabel('Countys')


plt.legend([line1, line2,line3,line4],barnames,loc=2)
plt.show()
```

```python
loaded_knn = pickle.load(open('data\knnsec.sav', 'rb'))
y_predKNN = loaded_knn.predict(X_test)
test_error_KNN = mean_squared_error(y_test,y_predKNN)
print(test_error_KNN)
```

```python
print("Predicted values : \n",y_predKNN)
print("Actual values : \n", np.array(y_test))



plt.figure(num=None, figsize=(30, 10), dpi=80, facecolor='w', edgecolor='k')

y_pos = np.arange(len(countys))
truelabels = np.array(y_test)
plt.xticks(y_pos, countys)

line1 = plt.bar(y_pos-0.47 , truelabels[:,0], color = 'black', width = 0.15 , label='label1')
line2 =plt.bar(y_pos-0.32 , y_predKNN[:,0], color = 'orange', width = 0.15 , label='label2')
plt.bar(y_pos-0.15, truelabels[:,1], color = 'black', width = 0.15 )
line3 = plt.bar(y_pos, y_predKNN[:,1], color = 'b', width = 0.15, label='label3')
plt.bar(y_pos+0.17 , truelabels[:,2], color = 'black', width = 0.15)
line4 = plt.bar(y_pos+0.32 , y_predKNN[:,2], color = 'g', width = 0.15 , label='label4')
plt.title('KNN predicts table')
plt.ylabel('Vote percentage(%)')
plt.xlabel('Countys')


plt.legend([line1, line2,line3,line4],barnames,loc=2)
plt.show()


```

KNN Regression mean squared error is less than 0.25 , this is verify my hypothesis. (Proof 2)

```python
loaded_NN = pickle.load(open('data/nnsec.sav', 'rb'))
y_predNN = loaded_NN.predict(X_test)
test_error_NN = mean_squared_error(y_test,y_predNN)
print(test_error_NN)
```

```python
print("Predicted values : \n",y_predNN)
print("Actual values : \n", np.array(y_test))

plt.figure(num=None, figsize=(30, 10), dpi=80, facecolor='w', edgecolor='k')

y_pos = np.arange(len(countys))
truelabels = np.array(y_test)
plt.xticks(y_pos, countys)

line1 = plt.bar(y_pos-0.47 , truelabels[:,0], color = 'black', width = 0.15 , label='label1')
line2 =plt.bar(y_pos-0.32 , y_predNN[:,0], color = 'orange', width = 0.15 , label='label2')
plt.bar(y_pos-0.15, truelabels[:,1], color = 'black', width = 0.15 )
line3 = plt.bar(y_pos, y_predNN[:,1], color = 'b', width = 0.15, label='label3')
plt.bar(y_pos+0.17 , truelabels[:,2], color = 'black', width = 0.15)
line4 = plt.bar(y_pos+0.32 , y_predNN[:,2], color = 'g', width = 0.15 , label='label4')
plt.title('NN predicts table')
plt.ylabel('Vote percentage(%)')
plt.xlabel('Countys')


plt.legend([line1, line2,line3,line4],barnames,loc=2)
plt.show()
```

Neural Network Regression mean squared error is less than 0.25 , this is verify my hypothesis. (Proof 3)
